> # Theory

## Terminologies/Cocepts
### Gradient Descent ($\nabla$)
Gradient descent is an __optimisation algorithm__ which is predominently used to train machine learning models and neural networks. Training data helps these models in learning, while the __loss function__ acts as a hallmark, oscilating and gauaging the accuracy and minimising the loss over each iteration (parameter update)
Note: Gradient produced by a linear 



### Loss Function ($J$)
Loss function is a measure of the __difference__ between __actual value__ $y$ and the __predicted value__ </span>&ycirc;</span>. It improves the model's efficiancy by providing feedback to the model, so that it can adjust the parameters (eg. weights, bias) to minimize the error and find the absolute local or absolute global minimum. The function will be iterated continously, moving along the direction of the steepest descent (or the -ve gradient) until the loss function is close to or at zero.



### Learning Rate ($\alpha$)
It is a tuning parameter, that determines the rate at which the model learns. In simpler terms it defines the adjustments in the weights of our network/model with respect to the loss gradient descent. It can also be understood as the step size of each iteration while moving towards the minimum loss value

### Predicting Values
__$\hat{y}$ (calc_ypred function)__<br>

The code provided implements the predicted value calculation for simple and multiple linear regression models. In simple linear regression, the predicted value is calculated using a single input variable, while in multiple linear regression, multiple input variables are used for predicting the target variable. The predicted value $\hat{y}$ is calculated using the following formulas:<br>

__Simple Linear Regression__: $\hat{y}_i = wx_i + b$<br>
__Multiple Linear Regression__: $\hat{y}_i = b + w_1(x_i)_1 + w_2(x_i)_2 + w_3(x_i)_3 + ..... w_n(x_i)_n $<br>

The purpose of these functions is to predict the outcome variable for a given set of input variables, given the weights and bias of the regression model.


### User defined function to calculate the predicted values (Simple Linear Regression) <br>
The code below implements a function to __calculate predicted values__ for __Simple Linear Regression__. The function "calc_ypred_slr" takes three parameters, X, weight, and bias. The parameter X is an array of input values, weight represents the slope of the regression line, and bias represents the intercept of the regression line.

To calculate predicted values, the function first initializes an empty array to store the predicted values. It then iterates over the length of the input array X using a for loop. Within the loop, it calculates the predicted value using the simple linear regression formula $\hat{y}_i = wx_i + b$ for each input value. The predicted value is then appended to the pred_values array. Finally, the function returns the array of predicted values.

`````python
#Simple Linear Regression
#Function to calculate y_pred (predicted values), based on current weight and bias 
def calc_ypred_slr(X,weight,bias):
    pred_values = []
    for i in range(len(X)):
        pred = (X[i] * weight) + bias
        pred_values.append(pred)
    return pred_values
`````

### User defined function to calculate the predicted values (Multiple Linear Regression) 
The code below represents the implementation of a function to __calculate predicted values__ for __Multiple Linear__ Regression. The function "calc_ypred" takes three parameters, X, weights, and bias. The parameter X is a two-dimensional array of input values, where each row represents a single observation, and each column represents a single feature. The parameter weights is an array of weights for each feature, and bias represents the intercept of the regression line.

The function first initializes an empty array to store the predicted values. It then iterates over the length of the input array X using a for loop. Within the loop, it initializes a variable pred to 0 and iterates over the length of the weights array using another for loop. Within this loop, it calculates the predicted value for each feature using the multiple linear regression formula $\hat{y}_i = b + w_1(x_i)_1 + w_2(x_i)_2 + w_3(x_i)_3 + ..... w_n(x_i)_n $. It then adds the calculated value to the variable pred. Finally, it appends the sum of predicted value and bias to the pred_values array. The function returns the array of predicted values.

`````python
#Multiple Linear Regression
#Function to calculate y_pred (predicted values), based on current weights and bias 
def calc_ypred(X,weights,bias):
    pred_values = []
    for i in range(len(X)):
        pred = 0
        for wt in range(len(weights)):
            pred = pred + (X[i][wt] * weights[wt])
        
        pred_values.append(pred + bias)               
                        
    return pred_values
`````

### User defined function to calculate the Loss Function ($J$) (MSE values) 
This code defines a user-defined function to calculate the Mean Squared Error (MSE) values between predicted and true values.

Mean Squared Error (MSE) is a commonly used loss function in regression problems. The MSE represents the average of squared differences between the predicted and actual values.

The function takes two parameters - the predicted values (y_pred) and the actual or true values (y_true). It calculates the total loss by iterating over each predicted value, finding the difference between the predicted and true value, squaring this difference, and adding it to a running sum. Finally, the total loss is divided by the number of predictions to obtain the average MSE.

The formula for the MSE used in this code is:

$J(w,b) = \frac{1}{2m}\sum{(\hat{y}_i-y_i)}^2$ $= \frac{1}{2m}\sum{(wx_i + b - y_i)}^2$

where $\hat{y}_i$ is the predicted value, $y_i$ is the true value, $w$ and $b$ are the weights and bias respectively, and $m$ is the number of samples.

Overall, this function is useful for evaluating the performance of a regression model during training, and is a key component in optimizing the model's weights and bias to minimize the loss.

`````python
#User defined function to calculate the MSE values
def loss_function_mse(y_pred,y_true):
    total_loss = 0 
    for i in range (0,len(y_pred)):
        total_loss += ((y_pred[i]- y_true[i])**2)
        
    return total_loss/(len(y_pred))
`````

### User defined function to calculate  Change in Bias  ($\delta{b}$)
In machine learning, it is often necessary to determine how much to adjust the parameters of a model in order to minimize a cost function. One way to do this is by calculating the gradient of the cost function with respect to the model parameters and using this gradient to update the parameters in the direction of steepest descent. The gradient can be decomposed into partial derivatives with respect to each parameter, which can be calculated using the chain rule of calculus.

One such parameter is the bias, which represents the intercept of the regression line in linear regression models. The change in bias, denoted as $\delta{w}$, can be calculated by taking the partial derivative of the cost function with respect to the bias. In the case of the mean squared error (MSE) cost function, the partial derivative with respect to the bias can be expressed as:

$\frac{\partial{j}}{\partial{b}} = \frac{1}{m}\sum{(\hat{y}_i-y_i)}$

Here, $\hat{y}_i$ represents the predicted value for the $i$th example and $y_i$ represents the corresponding true value. The function "change_in_bias" implements this calculation in Python, taking in an array of predicted values and an array of true values, and returning the average difference between the predicted and true values. This value can then be used to update the bias parameter in the model.

`````python
#Function to calculate the bias change(delta)
def change_in_bias(y_pred,y_true):
    total_bias = 0
    for i in range (0,len(y_pred)):
        total_bias += (y_pred[i] - y_true[i])
        
    return total_bias/len(y_pred)
`````

### User defined function to calculate  Change in Weight ($\delta{w}$) (Simple Linear Regression)
In the context of __simple linear regression__, the change in weight represents the adjustment made to the slope of the regression line to minimize the cost function. The cost function measures the difference between the predicted values and the actual values. In order to minimize the cost function, the change in weight must be calculated.

The formula to calculate the change in weight is given by:

$ \frac{\partial{j}}{\partial{w}} = \frac{1}{m}\sum{(\hat{y}_i-y_i)} * x_i $

Where $\hat{y}_i$ is the predicted value, $y_i$ is the actual value, $x_i$ is the input feature, and m is the total number of observations.

__Code Explanation__<br>

The code block below shows the implementation of a function to calculate the change in weight for simple linear regression. The function named "change_in_weights_slr" takes three parameters: y_pred, y_true, and X. The parameter y_pred is an array of predicted values, y_true is an array of actual values, and X is an array of input features.

The function first initializes a variable wj_change to 0. It then iterates over the length of input array X using a for loop. Within the loop, it calculates the change in weight using the above formula for each input feature. The calculated change is added to the variable wj_change. Finally, the function returns the average change in weight by dividing the sum of changes by the length of input array X. This returned value represents the slope adjustment required to minimize the cost function.

`````python
#Simple Linear Regression
#Function to calculate the change(delta) in weights
def change_in_weights_slr(y_pred,y_true,X):
    wj_change = 0
    for i in range (0,len(X)):
        wj_change += (y_pred[i]-y_true[i])*X[i]
        
    return wj_change/len(X)
`````

### User defined function to calculate  Change in Weight ($\delta{w}$) (Multiple Linear Regression)

In __multiple linear regression__, we use the change in weights ($\delta{w}$) to adjust the model parameters during training. The change in weights is calculated by taking the partial derivative of the cost function with respect to each weight, and then multiplying it by the learning rate. The partial derivative of the cost function with respect to each weight is calculated using the following formula:

__Change in Weight ($\delta{w}$)__<br> 
$ \frac{\partial{j}}{\partial{w}} = \frac{1}{m}\sum{(\hat{y}_i-y_i)} * x_i $<br>

Here, $\hat{y}_i$ is the predicted value, $y_i$ is the actual value, $x_i$ is the feature value, and $m$ is the number of training examples.

__Code Explanation__<br>
The `change_in_weights` function takes three parameters: `y_pred`, `y_true`, and `X`. `y_pred` is an array of predicted values, `y_true` is an array of true values, and `X` is a two-dimensional array of feature values. The function first initializes an empty array called `weight_change` to store the changes in weights. It then iterates over the length of the first dimension of `X` using a for loop. Within this loop, it initializes a variable called `wj_change` to store the change in weight for each feature. It then iterates over the length of `X` using another for loop. Within this loop, it calculates the change in weight for each feature using the formula mentioned above. Finally, the function appends the change in weight to the `weight_change` array and returns it.

`````python
#Multiple Linear Regression
#Function to calculate the change(delta) in weights
def change_in_weights(y_pred,y_true,X):
    weight_change = []
    for wght in range(0,len(X[0])):
        wj_change = 0
        for i in range (0,len(X)):
            wj_change += (y_pred[i]-y_true[i])*X[i][wght]
    
        weight_change.append(wj_change/len(X))
        
    return weight_change
`````

## User-defined function to perform Gradient Descent

Gradient Descent is an optimization algorithm used to find the optimal values of the parameters(weights and biases) for a given model that minimizes the loss function. The idea is to take steps in the opposite direction of the gradient of the loss function w.r.t the weights and biases. This process continues until we reach the global minimum of the loss function or when no improvement is observed after several iterations.

__Algorithm__

1. Set epoch (iteration) value -> epoch/n
2. Set Learning Rate -> 𝛼
3. Initiate Iteration Counter -> 0
4. Initiate Weight -> 0
5. Initiate Bias -> 0
6. Repeat Until iteration -> epoch
    - Y_pred = calc_ypred(X, weights, bias)
    - loss = loss_function(Y_pred, Y_actual)
    - Weight = Weight - (Learning Rate * change_in_weight(Y_pred, Y_actual, X))
    - Bias = Bias - (Learning Rate * change_in_bias(Y_pred, Y_actual))

__Code Explanation__

The function gradient_descent() takes the following parameters:<br>

- __X__: The input features as a numpy array
- __Y__: The target values as a numpy array
- __epochs__: The number of iterations for which the algorithm runs
- __learning_rate__: The step size at each iteration
- __slr__: A boolean value indicating whether the model is Simple Linear Regression or Multiple Linear Regression

The function first initializes the weights and bias values for the learner based on the value of slr. It then initializes an empty list loss_arr to store the loss, weight, and bias values after each epoch.

The function then runs a loop for the specified number of epochs. For each epoch, the function first predicts the Y values using the calc_ypred() or calc_ypred_slr() function based on the value of slr. It then calculates the mean squared error loss based on the predicted and actual Y values using the loss_function_mse() function.

The function then calculates the change in weights and biases using the change_in_weights() or change_in_weights_slr() and change_in_bias() functions. The weights and biases are then updated based on the learning rate and the calculated changes.

Finally, the function stores the loss, weight, and bias values for the current epoch in a dictionary and appends it to the loss_arr list. The function then returns the loss_arr list.

`````python
#User defined function to calculate gradient descent 
def gradient_descent(X,Y,epochs=100,learning_rate = 0.01,slr=False):
    #Initialising weights and bias values for the learner
    
    if slr==True:
        weights = 0
    else:
        weights = [0 for x in range(len(X[0]))]
    
    bias = 0

    loss_arr =[]
    
    for epoch in range(0,epochs):
        #Predicting Y values
        if slr==True:
            y_pred = calc_ypred_slr(X,weights,bias) 
            
        else:
            y_pred = calc_ypred(X,weights,bias) 
            
        #Calculating MSE loss based on predicted and actual value
        loss = loss_function_mse(y_pred,Y) 
        
        #Creating a dictionary to store loss, weight, bias values after each epoch
        loss_dict = {"Weights" : weights,
               "Bias" : bias,
               "Loss" : loss}
    
        #Calculating Change in weight/s
        if slr==True:
            new_weights_change = change_in_weights_slr(y_pred,Y,X)
            weights = weights - (learning_rate * new_weights_change)

        else:
            new_weights_change = change_in_weights(y_pred,Y,X)
        
            #Updating weight/s 
            weights = [(weights[n] - (learning_rate * new_weights_change[n])) 
                       for n in range(len(new_weights_change))]
        
        #Calculating Change in bias and Updating bias
        new_bias_change = change_in_bias(y_pred,Y)
        bias = bias - (learning_rate * new_bias_change)

        loss_arr.append(loss_dict)

    return loss_arr
`````

`````python
#Function to extract values in a dictionary based on the key
def extract_key(arr,key):
    arr_vals = []
    for x in arr:
        arr_vals.append(x[key])
    
    return arr_vals
`````

## User-defined functions to perform hyperparameters analysis

In machine learning, hyperparameters are the parameters of a model that are not learned through the training process. These parameters need to be set before the training process begins and have a significant impact on the performance of the model. Hyperparameters such as learning rate and number of epochs can greatly affect the convergence and accuracy of the model. In order to find the best combination of hyperparameters, we can perform a hyperparameter analysis by training the model with different values of hyperparameters and comparing their performance. In this section, we will discuss the user-defined functions to perform hyperparameters analysis for the gradient descent algorithm.

__Algorithm__

1. __parameter_analysis__ function<br>
    This user-defined function takes input as the data, epochs, learning rate, and a flag for a single linear regression. It returns a dictionary that contains the gradient descent models for different epochs and learning rates. It iterates through the learning rates and epochs and performs gradient descent on the data using the specified learning rate and number of epochs. It stores the resulting weights, bias, and loss values in a dictionary and appends it to a list. This list is then stored in another dictionary where the learning rate is the key and the list of dictionaries is the value.

2. __plot_parameter_analysis___ function<br>
    This user-defined function takes the output from the parameter_analysis function and a parameter to analyze (e.g. loss, weights, bias) as input. It plots the variation of the selected parameter with respect to the epochs for each learning rate. The function iterates through the learning rates and epochs, retrieves the stored values for the selected parameter, and plots them in a subplot for each epoch.

__Code explanation__<br>

The __parameter_analysis__ function takes four input arguments:<br>

- __X__: input features
- __Y__: target variable
- __epochs__: a list of integers representing the number of epochs to be used for training the model
- __learning_rate__: a list of floats representing the learning rates to be used for training the model
- __slr:__ a boolean flag to indicate whether the data is for simple linear regression or not.

The function initializes an empty dictionary gd_dict to store the gradient descent models. It iterates through the learning rates and epochs using nested for loops. For each combination of learning rate and epoch, the function calls the gradient_descent function with the specified parameters and stores the resulting weights, bias, and loss values in a dictionary dic_store. This dictionary is then appended to a list lr_i, which stores the values for all epochs for a particular learning rate. The list lr_i is then stored in another dictionary gd_dict, where the learning rate is the key and the list of dictionaries lr_i is the value. Finally, the gd_dict dictionary is returned.

The __plot_parameter_analysis__ function takes two input arguments:

- __gd_dict__: the output dictionary from the parameter_analysis function
- __prm__: a string representing the parameter to analyze (e.g. "Loss", "Weights", "Bias")
- __color__: a string representing the color to use for the plot (optional)

The function iterates through the learning rates in the gd_dict dictionary using a for loop. For each learning rate, it retrieves the corresponding list of dictionaries lr_i. It then iterates through the epochs for that learning rate using another for loop. For each epoch, it retrieves the stored values for the selected parameter prm from the dictionary lr_i[ep] and plots them in a subplot. The subplot title shows the number of epochs used for that iteration.  The function uses the plt module from the matplotlib library to create the plots. Finally, the function uses the ```plt.show``` to display all subplots together

`````python
#This user defined function returns a dictionary of gd models, based on various epochs and learning rates
def parameter_analysis(X,Y,epochs,learning_rate,slr= False):
    gd_dict = {}
    #Iterating throug the epochs and learning rates and performing gradient descent on the same
    for lr in learning_rates:
        lr_i = []
        for e in epochs:
            dic_store = {}
            store_dicts = gradient_descent(X,Y,e,lr, slr)
            dic_store["epochs"] = e
            dic_store["Loss"] = (extract_key(store_dicts,"Loss"))
            dic_store["Weights"]= (extract_key(store_dicts,"Weights"))
            dic_store["Bias"] = (extract_key(store_dicts,"Bias"))
            lr_i.append(dic_store)
        gd_dict[lr] = lr_i
    return gd_dict

#Plotting the hyperparameter analysis results
def plot_parameter_analysis(gd_dict,prm="Loss",color=None):
    print(("Comparing "+prm+" while Changing Epochs for given Learning Rate").center(100))
    keys_arr = list(gd_dict.keys())
    
    for lr in range(len(keys_arr)):
        plt.figure(figsize=(20,3))
        print("-"*110)
        print(("Learnign Rate : "+ str(keys_arr[lr])).center(100))
        lr_i = gd_dict[keys_arr[lr]]
    
        for ep in range(len(lr_i)):
            plt.subplot(1,len(lr_i),ep+1)
            text = str(lr_i[ep]['epochs']) + " Epochs (Iterations)"
            plt.title(text,color="purple")
            plt.plot(lr_i[ep][prm],linewidth=4.5,color=color)
            plt.xlabel("Epochs",color="Brown")
            plt.ylabel(prm,color="Brown")


        plt.tight_layout(pad=1)
        plt.show()

    print("-"*110)
`````

The following code implements the gradient descent algorithm on a simple linear regression model. The aim of this implementation is to calculate the optimal values of the slope and the intercept, which minimize the cost function. The cost function measures the difference between the predicted and actual values of the dependent variable. The gradient descent algorithm adjusts the values of the slope and intercept iteratively to minimize the cost function.

It aims to minimize the sum of the squared errors between predicted and true values by finding the optimal coefficients (weights).
