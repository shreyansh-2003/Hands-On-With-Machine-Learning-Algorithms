> # Theory

## Finding Optimal 'k' clusters

### 1. Elbow Method
The elbow method, is a graphical representation of finding the _optimal 'k' (number of clusters)_ while conducting a clustering Analysis

__Axises__

- __X Axis:__ Number of Clusters (k)<br>
- __Y_Axis:__ Variation (variability) in each cluster / WCSS (Within Cluster Sum of Squares)<br>

__Within Cluster Sum Of Squares (WCSS) Formula__

Distance between $d_i$ (ith data point) , $c_i$ (ith data point's centroid)

$WCSS$ = {\sum{(d_i,c_i)^2}}$ 



### 2. Silhouette Method
The silhouette Method is also a method to find the _optimal number of clusters_ while aiding interpretation and validation of consistency within clusters of data.<br>
The silhouette method computes __silhouette coefficients__ of each point that measure how much a point is similar to its own cluster compared to other clusters.

__Formula (Silhoette Coefficient)__

Computing s(i)  silhouette coefficient or ith point using below mentioned formula.

$s(i) = \frac{b(i) - a(i)}{\max\{a(i),b(i)\}}$

1. __Compute a(i)__: The average distance of that point with all other points in the same clusters.
2. __Compute b(i)__: The average distance of that point with all the points in the closest cluster to its cluster.

## Hierarchical Clustering (Linkages, Denrogram)

__Linkage__: The method in which the distance between two datapoints is calculated and the method used for deciding which two data points to connect in Hierarchical Clustering

__Dendrogram__: A herarchichal tree and method that plots and depicts hierarchical Clustering.



__Types of linkages__:

1. __Single linkage:__ This method tends to create long, stretched-out clusters. It can also result in clusters being merged incorrectly, leading to poor results.

2. __Complete linkage:__ This method tends to create tight, compact clusters. It is less sensitive to outliers than single linkage and tends to perform well in practice.

3. __Average linkage:__ This method is a compromise between single and complete linkage. It tends to create well-balanced clusters that are not too stretched out.

4. __Weighted linkage:__ This method is similar to average linkage, but it gives more weight to larger clusters. This can help to prevent small clusters from being merged too early.

5. __Centroid linkage:__ This method is based on the distance between cluster centroids. It tends to create balanced clusters that are not too stretched out. However, it can be sensitive to outliers.

6. __Median linkage:__ This method is similar to centroid linkage, but it uses the median distance between points instead of the mean. It tends to be more robust to outliers than centroid linkage.

7. __Ward linkage:__ This method minimizes the sum of squared differences within each cluster. It tends to create tight, compact clusters and is often used for hierarchical clustering.

# K-Means Algorithm
__Basic Idea / Blueprint / Psudocode__

1. Chose the __number of clusters 'k'__.
2. Chose 'k' random data points in the dataset as __centroids__.
3. Iterate_ through the data points.
    - Chose a data point. 
    - Calculate the euclidian distance between it and all the centroids.
    - Assign the datapoint to the the nearest cluster centroid.
4. Update the centroid of each respective cluster as the mean of all the data points in that cluster.
5. Repeat Step 3 and 4 until the cluster assignment no more changes or until maximum iterations are reached.

````python
#K_means class for performing KMeans clustering algorithm
class K_means:
    
    #Attributes/Constructor
    def __init__(self, dataset, n_clusters = 3, max_iter = 500, seed_value = None ):

        if seed_value != None:
            random.seed(seed_value)
        else:
            pass
        
        self.dataset = dataset
        self.n_clusters = n_clusters
        self.max_iter = max_iter 

        #Centroids mapped to each cluster
        self.centroids = {}
        #Clusters for each data point in array form
        self.clusters = []


    #User defined function to compute and predict the k-means clusters
    def predict(self):
        
        #Step 1 --> 'k' number of clusters, as set
        k = self.n_clusters
    
        #Step 2 --> Initialising 'k' random data points as centroids
        centroid_values = random.sample(self.dataset , k)
        #Creating a dictionary of all the new centroid values mapped with their cluster
        for i in range(0,k):
            self.centroids[i] = centroid_values[i] 
    
        #Step 5 --> Repeating step 3 and 4 until either of the following conditions aren't met
        # Condtion 1: Max Iterations are Reached 
        # Conditon 2: Cluster Assignments are not changing 
        iterations = 0
        while iterations != self.max_iter:
            
            #Step 3 --> Iterating through all the data points and generating and assigning clusters
            new_clusters = self.perform_kmeans()
            
            #Step 4 ---> Update the centroid of each respective cluster as the mean of all the data points
            new_centroids = self.calculate_new_centroid(new_clusters)
            
            if new_centroids == self.centroids:
                break

            else: 
                self.clusters = new_clusters
                self.centroids = new_centroids
                iterations = iterations + 1 
        
        return iterations, self.clusters, self.centroids

    
    #User-defined function to calculate k-means clustering 
    def perform_kmeans(self):
        #Creating the array that will store mapped clusters
        clusters_arr = []
        #Iterating through all the data points
        for data_point in self.dataset:
            eucc_distacnes = self.calculate_euclidean_distance(data_point) #Calculate euclidean distance
            cluster_i = self.chose_centroid(eucc_distacnes) #Chose the closest cluster
            clusters_arr.append(cluster_i)

        return clusters_arr
    
    
    #User defined function to calculate the euclidean distance from each cluster 
    def calculate_euclidean_distance(self,data_point):
        eucc_distances = {}
        #Iterating through each centroid
        for c in range(0,len(self.centroids.keys())):
            centroid_i = self.centroids[c]
            eucc_dist_sum = 0
            
            #Calculating the euclidean distance wrt iterated n cenroid
            for i in range(0,len(centroid_i)):
                eucc_dist_sum += ((data_point[i] - centroid_i[i])**2)

            #Square root of summated data point to centroid distance
            eucc_dist_i = (eucc_dist_sum)**(0.5)

            #Adding the distance to the dictionary
            eucc_distances[c] = eucc_dist_i
        
        return eucc_distances
        
    
    #User-defined funcition to chose the centroid which has the lowest distance to the centroid
    def chose_centroid(self,eucc_distances):
        for x in range(0,len(eucc_distances.keys())):
            if eucc_distances[x] == min(eucc_distances.values()):
                return x
            else:
                continue
                
    #User Defined function to calculate the new centroids based on current clusters
    def calculate_new_centroid(self,clusters):
        cluster_dict = {}
        for x in range(0,len(clusters)):
            if clusters[x] not in cluster_dict.keys():
                cluster_dict[clusters[x]] = []
                cluster_dict[clusters[x]].append(self.dataset[x])
            else:
                cluster_dict[clusters[x]].append(self.dataset[x])
                
        new_centroids = {}
        
        for cluster_i in cluster_dict.keys():

            
            new_cluster_centroid = []
            
            for feature in range(0,len(cluster_dict[cluster_i][0])):
                feature_total = 0
                
                for data_point in range(0,len(cluster_dict[cluster_i])):
                    feature_total +=  cluster_dict[cluster_i][data_point][feature]

                new_cluster_centroid.append(feature_total/len(cluster_dict[cluster_i]))

            new_centroids[cluster_i] = new_cluster_centroid
            
        return new_centroids

    
    #User Defined function to calculate WCSS (within cluster sum of squares)
    def calculate_wcss(self):
        # within cluster sum of squares initialised to 0
        wcss = 0

        # for each centroid/cluster
        for cluster in range(0,len(self.centroids.keys())):
            centroid_i = self.centroids[cluster]
            cluster_data_points = [self.dataset[i] 
                                   for i in range(len(self.clusters)) 
                                   if self.clusters[i] == cluster]
            cluster_sum_of_squares = 0

            # for each datapoint in the cluster
            for data_point in cluster_data_points:
                data_point_wcss = 0

                # for each feature
                for feature in range(len(data_point)):
                    feature_centroid_i = centroid_i[feature]
                    eucc_dist = (data_point[feature] - feature_centroid_i )**2
                    data_point_wcss += eucc_dist

                cluster_sum_of_squares += data_point_wcss

            wcss += cluster_sum_of_squares

        return wcss 
    
    
    #User defined function to plot elbow plot for various K's 
    def elbow_plot(self,k_limit,x,y):
        wcss_k = []
        for k in range(1,k_limit+1):
            km = K_means(self.dataset, k, max_iter = 500, seed_value = 1)
            iters_i, cl_i, centroids_i = km.predict()
            wcss_k_i = km.calculate_wcss()
            wcss_k.append(wcss_k_i)
            
        plt.figure(figsize=(x,y))
        plt.grid()
        plt.title("Elbow Plot for Selection of Optimal K clusters", c = "Purple",fontsize = 30)
        plt.plot(range(1,k_limit+1,1),wcss_k,"--",color = 'black',linewidth=4)
        plt.xlabel("K",c ="Brown",fontsize = 20)
        plt.ylabel("WCSS (Within Cluster Sum Of Squares)", c ="Brown",fontsize = 20)
        
        return plt.show()
    
    # User Defined function to compute the silhouette scores for different cluster sizes
    def silhouette_comp_chart(self,k_limit,x,y):
        silhouette_scores = []
        for k in range(2, k_limit+1):
            km = KMeans(n_clusters=k, max_iter=1000, 
                        n_init = 1, random_state = 42, 
                       init = "k-means++")
            cl_i = km.fit_predict(X)
            scores = silhouette_score(X, cl_i)
            silhouette_scores.append(scores)

        # Plot the silhouette scores as a line chart
        plt.figure(figsize=(x,y))
        plt.grid()
        plt.title("Silhouette Score Comparison Chart for Optimal K clusters ",
                  c="purple", 
                  fontsize=30)
        plt.plot(range(2, k_limit+1,1), silhouette_scores, linewidth=3, marker="x", color="black")
        plt.xlabel("Number of clusters", c="brown", fontsize=20)
        plt.ylabel("Average silhouette score", c="brown", fontsize=20)
        plt.grid()

        print("Optimum number of clusters, according to the silhouette scores are", 
              silhouette_scores.index(max(silhouette_scores))+2)

        return plt.show()
````

__Manually Searching for best hyperparameters (Silhouette Score)__

````python
#Creating the evaluation dataframe
kmeans_eval = pd.DataFrame(columns=['k','init(method)',
                                    'max_iterations',
                                    'n_init',
                                    'silhouette_score'])

#Creating an array of all hyperparameters to run through
n_clusters_arr = [2, 3, 4, 5, 6, 7, 8, 9, 10]
init_arr = ['k-means++', 'random']
max_iter_arr = [100, 500, 1000]
n_init_arr = [1,5,10]

#Iterating through the hyperparameters
for k in n_clusters_arr:
    for mtd in init_arr:
        for iters in max_iter_arr:
            for n_init_i in n_init_arr:
                
                #performing k means with each different permutation of parameters
                km_i= KMeans(n_clusters = k,
                           init = mtd,
                           max_iter = iters,
                        n_init = n_init_i, 
                         random_state = 42)
            
                cluster_i = km_i.fit_predict(X)
                
                #calculating silhouette score
                score = calculate_silhouette_score(cluster_i, X)
                
                #Adding evaluation to the dataset
                eval_i = {'k' : k ,
                 'init(method)' : mtd,
                 'max_iterations' : iters,
                 'n_init' : n_init_i,
                'silhouette_score' : score}

                kmeans_eval = kmeans_eval.append(eval_i, ignore_index = True)
````

````python
#User defined function to compute the k-means clustering algorithm 
def kmeans_algo(dataset,n_clusters,max_iter,seed_value=None):
    
    if seed_value!=None:
        random.seed(seed_value)
    else:
        pass
    
    #Step 1 --> 'k' number of clusters, as set
    k = n_clusters 
    
    #Step 2 --> Initialising 'k' random data points as centroids
    centroid_values = random.sample(dataset,k)
    #Creating a dictionary of all the new centroid values mapped with their cluster
    centroids = {}
    for i in range(1,k+1):
        centroids[i] = centroid_values[i-1] 
    
    
    #Step 5 --> Repeating step 3 and 4 until either of the following conditions aren't met
    # Condtion 1: Max Iterations are Reached 
    # Conditon 2: Cluster Assignments are not changing 
    iterations = 0
    clusters = []
    while iterations!=max_iter:
        #Step 3 --> Iterating through all the data points and generating centroids
        new_clusters = perform_kmeans(dataset,k,centroids)
        #Step 4 ---> Update the centroid of each respective cluster as the mean of all the data points
        new_centroids = calculate_new_centroid(dataset,new_clusters)
    
        if new_centroids == centroids:
            break
        
        else: 
            clusters = new_clusters
            centroids = new_centroids
            iterations = iterations + 1 
        
    print(iterations)

    return clusters, centroids

#User-defined function to calculate k-means clustering 
def perform_kmeans(dataset,k,centroids):
    #Creating the array that will store mapped clusters
    clusters = []
    #Iterating through all the data points
    for data_point in dataset:
        eucc_distacnes = calculate_euclidean_distance(data_point, centroids)
        cluster_num = chose_centroid(eucc_distacnes)
        clusters.append(cluster_num)
    
    return clusters

#User defined function to calculate the euclidean distance from each cluster 
def calculate_euclidean_distance(data_point, centroids):
    eucc_distances = {}
    #Iterating through each centroid
    for c in range(1,len(centroids)+1):
        centroid_i = centroids[c]
        eucc_dist_sum = 0
        #Calculating the euclidean distance wrt iterated n cenroid
        for i in range(0,len(centroid_i)):
            eucc_dist_sum += (data_point[i] - centroid_i[i])**2
        
        #Square root of summated data point to centroid distance
        eucc_dist_i = eucc_dist_sum**(1/2)
        
        #Adding the distance to the dictionary
        eucc_distances[c] = eucc_dist_i
        
    return eucc_distances

#User-defined funcition to chose the centroid which has the lowest distance to the centroid
def chose_centroid(eucc_distacnes):
    for x in eucc_distacnes:
        if eucc_distacnes[x] == min(eucc_distacnes.values()):
            return x
        else:
            continue
            
#User Defined function to calculate the new centroids based on current clusters
def calculate_new_centroid(dataset,clusters):
    #Clusterwise segregations of datapoints
    cluster_dict = {}
    for x in range(0,len(clusters)):
        if clusters[x] not in cluster_dict:
            cluster_dict[clusters[x]] = []
            cluster_dict[clusters[x]].append(dataset[x])
        else:
            cluster_dict[clusters[x]].append(dataset[x])
            
    new_centroids = {}
    for cluster_i in cluster_dict.keys():
        new_cluster_centroid = []
        for feature in range(0,len(cluster_dict[cluster_i][0])):
            feature_total = 0
            for data_point in range(0,len(cluster_dict[cluster_i])):
                feature_total +=  cluster_dict[cluster_i][data_point][feature]
                
            new_cluster_centroid.append(round(feature_total/len(cluster_dict[cluster_i]),10))
        
        new_centroids[cluster_i] = new_cluster_centroid
    
    return new_centroids
````
